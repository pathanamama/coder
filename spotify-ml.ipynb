```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
df=pd.read_csv('/kaggle/input/top-spotify-songs-2023/spotify-2023.csv',  encoding='latin-1')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python

df['track_name'] = pd.to_numeric(df['track_name'], errors='coerce', downcast='integer')
df.dtypes
```

```python
cclum=[]
for i in df.columns:
  if df [i].dtypes=='object':
    cclum.append(i)
print(cclum)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df[cclum]=df[cclum].apply(le.fit_transform)
df.head(5)
```

```python
x=df.drop(['in_spotify_charts'],axis=1)
y=df['in_spotify_charts']
x=x.to_numpy()
x.shape
```

```python
y=df.in_spotify_charts
```

```python
features =['artist(s)_name','artist_count','released_year','released_month','released_day','in_spotify_playlists','streams']
x=df[features]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y , random_state=1 )
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
s_model = DecisionTreeRegressor(random_state=1)
```

```python
s_model.fit(X_train, y_train)
```

```python
# Make validation predictions and calculate mean absolute error
val_predictions = s_model.predict(X_test)
val_mae = mean_absolute_error(val_predictions, y_test)
print("Validation MAE when not specifying max_leaf_nodes: {:,.0f}".format(val_mae))
```

```python
# Using best value for max_leaf_nodes
s_model = DecisionTreeRegressor(max_leaf_nodes=100, random_state=1)
s_model.fit(X_train, y_train)
val_predictions = s_model.predict(X_test)
val_mae = mean_absolute_error(val_predictions, y_test)
print("Validation MAE for best value of max_leaf_nodes: {:,.0f}".format(val_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Define the model. Set random_state to 1
rf_model = RandomForestRegressor(random_state=1)

# fit your model
rf_model.fit(X_train, y_train)

# Calculate the mean absolute error of your Random Forest model on the validation data
rf_val_predictions = rf_model.predict(X_test)
rf_val_mae = mean_absolute_error(y_test, rf_val_predictions)
print(rf_val_mae)
```

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import confusion_matrix,classification_report
names=["SVM","NB","KNN","DT","RF","ET"]
clfs=[SVC(),GaussianNB(),KNeighborsClassifier(),DecisionTreeClassifier(),RandomForestClassifier(),ExtraTreesClassifier()]
CLFF=[]
```

```python
for name,clf in zip(names,clfs):
    print(name)
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    CLFF.append(clf)
    print(confusion_matrix(y_test,y_pred))
    print(classification_report(y_test,y_pred, zero_division=0))
    print('-----------------------------------------------')
```
